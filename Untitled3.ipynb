```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np

#باز کزدن دیتاست
df = pd.read_csv(r"/data.csv")

#نمایش تمام ردیف ها
pd.set_option('display.max.rows' , 10)

df
```

```python
#NaN حذف برخی ردیف ها با مقادیر

df = df.dropna(subset=['Quantity', 'Total'], how='all')
df = df.dropna(subset=['Product'], how='all')
```

```python
#پر کردن رکورد های قابل بازتولید با ستون های کناری

df.loc[df['Quantity'].isna(), 'Quantity'] = df['Total'] / df['Price']
df.loc[df['Price'].isna(), 'Price'] = df['Total'] / df['Quantity']
df.loc[df['Total'].isna(), 'Total'] = df['Price'] * df['Quantity']
```

```python

#نگه داشتن ردیف ها با مقادیر مثبت
df = df[df['Price'] >= 0]

#ریست کردن ایندکس گذاری چون به علت حذف برخی ردیف ها بهم ریخته است
df = df.reset_index(drop=True)

#برای احتیاط یک دیتاست میسازیم که بعدا از تغییرات پشیمان شدیم
df1=df
df
```

```python
#ساختن یک ستون جدید از روی ستون تاریخ برای ماه

df['Month'] = df['Date'].str[5:7]

df
```

```python
#بدست آوردن مجموع فروش ماهانه

monthly_sales = df.groupby('Month')['Total'].sum().reset_index()

monthly_sales
```

```python
#رسم نمودار برای روند فروش کل ماهیانه

import matplotlib.pyplot as plt

plt.figure(figsize=(7, 3))
plt.plot(monthly_sales['Month'], monthly_sales['Total'], marker='o', linestyle='-', color='blue')


plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#پیدا کردن محصول با بیشترین تعداد فروش

grouped = df.groupby('Product')['Quantity'].sum().reset_index()
grouped = grouped.sort_values(by='Quantity', ascending=False)
print(grouped)
```

```python
#پیدا کردن محصول با بیشترین قیمت فروش


grouped1 = df.groupby('Product')['Price'].sum().reset_index()
grouped1 = grouped1.sort_values(by='Price', ascending=False)
print(grouped1)
```

```python
# بررسی ضریب همبستگی بین Price و Quantity

correlation = df['Price'].corr(df['Quantity'])
print("ضریب همبستگی:", correlation)
```

```python
# پیدا کردن روز هایی با بیشترین فروش کل


grouped2 = df.groupby('Date')['Total'].sum().reset_index()
grouped2 = grouped2.sort_values(by='Total', ascending=False)
print(grouped2)
```

```python
#پیدا کردن محصول با کمترین مقدار فروش کل 


grouped3 = df.groupby('Product')['Total'].sum().reset_index()
grouped3 = grouped3.sort_values(by='Total', ascending=False)
print(grouped3)
```
